```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
a=np.random.normal((3,5,4),(1.25,1,1.25),(100,3))
a
fig,ax=plt.subplots()
VP=ax.boxplot(a,positions=[2,4,6],widths=1.0,showmeans=True,showfliers=False,
              medianprops={"color":"blue","linewidth":0.5},
              boxprops={"color":"C0","color":"orange","linewidth":0.5},
             whiskerprops={"color":"g","linewidth":1.5},
             capprops={"color":"k","linewidth":1.5})
```

```python
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
```

```python
x=np.linspace(0,5,50)

y=np.linspace(0,5,40)

X,Y=np.meshgrid(x,y)
print(X.shape)
print(Y.shape)
Z=f(X,Y)
print(Z.shape)


plt.contour(X,Y,Z,20,cmap='RdGy')
contours=plt.contour(X,Y,Z,20,cmap='RdGy')
#red were minima
#black were maxima
plt.clabel(contours,inline=True,fontsize=8)
plt.colorbar()

  
```
