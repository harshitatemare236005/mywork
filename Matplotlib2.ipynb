```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
d1=np.random.randn(1000)
d2=np.random.normal(0,0.8,1000)
d3=np.random.normal(-2,1,1000)
d4=np.random.normal(3,2,1000)
plt.hist(d1)
plt.hist(d1,bins=20,alpha=0.3,histtype='stepfilled')
```

```python
plt.hist(d2,bins=20,alpha=0.3,histtype='stepfilled')
plt.hist(d3,bins=20,alpha=0.3,histtype='stepfilled')
plt.hist(d4,bins=20,alpha=0.3,histtype='stepfilled')
```

```python
dict={'A':np.random.randn(365).cumsum(0),
      'B':np.random.randn(365).cumsum(0)+20,
      'C':np.random.randn(365).cumsum(0)-20
     }
```

```python
df=pd.DataFrame(dict,index=pd.date_range('1/1/2021',periods=365))
df.head()
```

```python
df.plot()
```

```python
df.plot('A','B',kind='scatter',color='red')
```

```python
scAB=df.plot.scatter('A','B',c=df['B'],colormap='viridis')
scAB.set_aspect('equal')
```

```python
df.plot.box()
```
