```python
# Sample data
dict1={'A':np.random.randn(365).cumsum(0),
     'B':np.random.randn(365).cumsum(0)+20,
     'C':np.random.randn(365).cumsum(0)-20}
sizes = [100, 200, 300, 400, 500]  # Bubble sizes

# Plotting the bubble chart
plt.figure(figsize=(8, 6))
plt.scatter(dict1["A"], dict1["B"], s=dict1["C"]*10, alpha=0.5, c='blue', edgecolors='black')  # 's' sets bubble sizes, 'alpha' sets transparency
plt.title('Bubble Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Customizing legend
bubble_sizes = [100, 200, 300, 400, 500]
bubble_labels = ['100', '200', '300', '400', '500']
bubble_legend = plt.legend(handles=[plt.scatter([], [], s=size, edgecolors='black', label=label, alpha=0.5) for size, label in zip(bubble_sizes, bubble_labels)],
                           title='Bubble Sizes', loc='best')
plt.gca().add_artist(bubble_legend)

plt.grid(True)
plt.show()
```

```python
# Generating random data for scatter plot
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

# Creating scatter plot with histograms
plt.figure(figsize=(8, 6))

# Scatter plot
plt.subplot(221)
plt.scatter(x, y, alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Histogram along the x-axis
plt.subplot(222)
plt.hist(x, bins=20, color='skyblue', alpha=0.7, orientation='vertical')
plt.title('Histogram - X-axis')
plt.xlabel('X-axis')
plt.ylabel('Frequency')

# Histogram along the y-axis
plt.subplot(223)
plt.hist(y, bins=20, color='salmon', alpha=0.7, orientation='horizontal')
plt.title('Histogram - Y-axis')
plt.xlabel('Frequency')
plt.ylabel('Y-axis')

# 2D Histogram
plt.subplot(224)
plt.hist2d(x, y, bins=20, cmap='Blues')
plt.colorbar(label='Frequency')
plt.title('2D Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
