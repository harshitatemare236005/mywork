```python
from tkinter import *
from PIL import Image, ImageTk
import os

root = Tk()
root.geometry('350x200')
files = os.listdir()

# For all types of images
x_cord = 0
y_cord = 0

for file in files:
    if file.endswith(".jpeg") or file.endswith(".png") or file.endswith(".jpg"):
        image = Image.open(file)
        resize = image.resize((100, 100))
        photo = ImageTk.PhotoImage(resize)
        lab = Label(image=photo)
        lab.place(x=x_cord, y=y_cord)
        lab.image = photo
        x_cord += 100  # Increment x-coordinate instead of y-coordinate

root.mainloop()
```

```python
class GUI(Tk):
    def __init__(self):
        super().__init__()
        self.geometry("550x500")
        self.title("Regression Analysis")
        
    def status(self):
        self.var = StringVar()
        self.var.set(" READY ")
        self.statusbar = Label(self,textvariable = self.var,relief = SUNKEN, anchor = "w")
        self.statusbar.pack(side = BOTTOM, fill = X)
        
    def Heading(self):
        Label(self,text = "Regression Analysis on iris Data",font = "Aerial 20 bold",relief = SUNKEN).pack(fill = X)
        
    def hii():
        print("Hii!!")
        
    def click(self):
        print("  !!!  ")
        
    def feature_Buttons(self,features):
        f = Frame(self)
        for i in features:
            Button(f,text = i,font = "Aerial 15 ",command = self.click).pack(side = LEFT,padx = 10,pady = 20)
        f.pack()
    
    def create_button(self,name,func):
        Button(self,text = name,command = func).pack()
        
    def show(self):
        self.mainloop()
```
